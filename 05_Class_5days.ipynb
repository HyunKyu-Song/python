## 데이터 프레임 생성하기

```python
import pandas as pd

data = {"이름" : ["사쿠라", "김채원", "허윤진", "카즈하", "홍은채"],
       "출생년도" : [1998, 2000, 2001, 2003, 2006],
       "점수" : [1.5, 1.7, 3.6, 2.4, 2.9]}

df = pd.DataFrame(data)
df
```
